```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import metrics

names = ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'Class']
datasets = pd.read_csv('./data/iris.data', names=names)
datasets.head()
```

```python
datasets.Class.unique()
```

```python
datasets.tail()
```

```python
x = datasets.iloc[:, :-1].values
y = datasets.iloc[:,4].values
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split

#train_set, test_set 분리
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

#데이터를 평균이 0, 표준편차 1이 되도록 변환(각 데이터셋의 크기가 뒤죽박죽인 경우 많기 때문)
s = StandardScaler()
x_train = s.fit_transform(x_train)
x_test = s.fit_transform(x_test)
print(x_train[:5], x_test[:5])
```

```python
from sklearn.neighbors import KNeighborsClassifier

#k=50개로 분류
knn = KNeighborsClassifier(n_neighbors=50)
knn.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = knn.predict(x_test)
print(f"정확도: {accuracy_score(y_test, y_pred)}")
```

```python
#최적의 k 찾기
k = 100
acc_array = np.zeros(k)

for k in np.arange(1, k+1, 1): #1부터 k까지 값을 취함
    classifier = KNeighborsClassifier(n_neighbors=k).fit(x_train, y_train)
    y_pred = classifier.predict(x_test) #x_test 데이터로 y값 예측
    acc = metrics.accuracy_score(y_test, y_pred) #y_test값과 y_pred값으로 정확도 산출
    acc_array[k-1] = acc #acc_array에 for문 번호에 해당하는 acc값 입력
    
max_acc = np.amax(acc_array)
acc_list = list(acc_array)
k = acc_list.index(max_acc) #최고 정확도 나온 값을 최적의 k로 함
print(acc_list)
print(k)
print("정확도", max_acc, "으로 최적의 k는", k+1, "입니다")
```
